```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
# Fonction a minimiser
fc = lambda x,y: (x**2) + (25*y**2)
# Calcul des dérivées partielles
D_x = lambda x : 2*x
D_y = lambda y : 50*y
# Initialisation des variables
x = 0.8
y = 0.8
# Pas d'apprentissage 
lr = 0.035
print ("Valeur initiale de x et y")
print (fc(x,y))
print ("Nouvelles valeurs calculées lors de l'entrainement")

x_values = []
y_values = []
x_values.append(x)
y_values.append(y)
# epoch : période de minimisation
for epoch in range(0,50):
    # Calcul des gradients
    G_x = D_x(x)
    G_y = D_y(y)
    # Appliquer la descente de gradients
    x = x - lr*G_x
    y = y - lr*G_y
    # Vérifier la nouvelle valeur
    print ("Fc= %s" % (fc(x,y)))
    x_values.append(x)
    y_values.append(y)

print(f'Valeurs finales de x,y')
print(f'x = {x}')
print(f'y = {y}')

```

**Evolution de x,y**

```python
x = np.linspace(-1,1)
y = np.linspace(-1,1)
x,y = np.meshgrid(x,y)
z = 2*x**2+(25*y**2)
levels = [1,6,9,15,20,30,50]
cs=plt.contour(x,y,z,levels)
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x_values, y_values)
plt.show()
```
