Objectif = visualiser l'évolution de l'accuracy, la loss, la val accuracy et la val loss durant l'entraînement pour identifier un potentiel overfitting/underfitting

```python
import matplotlib.pyplot as plt

file_name = 'train.txt'
with open('train.txt','r', encoding="utf-8") as fin:
    ep = []
    acc = []
    loss = []
    val_acc = []
    val_loss = []
    for line in fin:
        line = line.strip().split()
        if 'Epoch' in line:
            ep.append(float(line[1]))
        if 'Accuracy' in line:
            acc.append(float(line[2]))
        if 'Loss' in line:
            loss.append(float(line[2]))
        if 'Val' in line and 'accuracy' in line:
            val_acc.append(float(line[3]))
        if 'Val' in line and 'loss' in line:
            val_loss.append(float(line[3]))
```

```python
plt.plot(ep, acc, label='Training Accuracy')
plt.plot(ep, val_acc, label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
plt.plot(ep, loss, label='Training Loss')
plt.plot(ep, val_loss, label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
